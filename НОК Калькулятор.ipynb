–°—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–∞—è –æ—à–∏–±–∫–∞ (Mean Squared Error) ‚Äì –°—Ä–µ–¥–Ω–µ–µ –∞—Ä–∏—Ñ–º–µ—Ç–∏—á–µ—Å–∫–æ–µ (Mean) –∫–≤–∞–¥—Ä–∞—Ç–æ–≤ —Ä–∞–∑–Ω–æ—Å—Ç–µ–π –º–µ–∂–¥—É –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–Ω—ã–º–∏ –∏ —Ä–µ–∞–ª—å–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –ú–æ–¥–µ–ª–∏ (Model) –ú–∞—à–∏–Ω–Ω–æ–≥–æ –æ–±—É—á–µ–Ω–∏—è (ML):

–í –æ—Ç–ª–∏—á–∏–µ –æ—Ç —Å—Ä–µ–¥–Ω–µ–∫–≤–∞–¥—Ä–∞—Ç–∏—á–µ—Å–∫–∏—Ö –æ—à–∏–±–æ–∫, –≥–¥–µ –∏—Å–ø–æ–ª—å–∑—É–µ—Ç—Å—è –∫–≤–∞–¥—Ä–∞—Ç —Ä–∞–∑–Ω–æ—Å—Ç–∏, MAE —è–≤–ª—è–µ—Ç—Å—è –ª–∏–Ω–µ–π–Ω–æ–π –æ—Ü–µ–Ω–∫–æ–π, –ø–æ—ç—Ç–æ–º—É –≤–µ—Å —Ä–∞–∑–Ω–æ—Å—Ç–µ–π –æ–¥–∏–Ω–∞–∫–æ–≤ –Ω–µ–∑–∞–≤–∏—Å–∏–º–æ –æ—Ç –¥–∏–∞–ø–∞–∑–æ–Ω–∞. –ù–∞–ø—Ä–∏–º–µ—Ä –æ—à–∏–±–∫–∞ –º–µ–∂–¥—É 2 –∏ 4 –±—É–¥–µ—Ç –≤ –¥–≤–∞ —Ä–∞–∑–∞ –º–µ–Ω—å—à–µ, —á–µ–º –º–µ–∂–¥—É 4 –∏ 8, —Ç.–µ. 2 –∏ 4 —Å–æ–æ—Ç–≤–µ—Ç—Å—Ç–≤–µ–Ω–Ω–æ (–≤ —Ç–æ –≤—Ä–µ–º—è –∫–∞–∫ –ø—Ä–∏ –∫–≤–∞–¥—Ä–∞—Ç–∏—á–Ω–æ–π –æ—Ü–µ–Ω–∫–µ 4 –∏ 16, —Ç.–µ. –≤ 4 —Ä–∞–∑–∞ –º–µ–Ω—å—à–µ).

# –ó–∞–¥–∞–Ω–∏–µ 17: Z-test

```python
from statsmodels.stats.weightstats import ztest


# Z-—Ç–µ—Å—Ç:
#   data - –≤—ã–±–æ—Ä–∫–∞
#   value - –∑–Ω–∞—á–µ–Ω–∏–µ –Ω—É–ª–µ–≤–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã
#   alternative - —Ç–∏–ø –∞–ª—å—Ç–µ—Ä–Ω–∞—Ç–∏–≤–Ω–æ–π –≥–∏–ø–æ—Ç–µ–∑—ã: 'two-sided', 'larger', 'smaller'

a = [103.01, 101.99, 105.21, 106.80, 112.70, 106.13, 110.48, 109.26, 100.44, 100.28]
b = [107.38, 106.31, 106.00, 105.27, 105.27, 104.66, 103.70, 105.07, 105.12, 104.74]

# –¢—É—Ç –Ω–∞–º –Ω–∞–¥–æ –ø—Ä–∏–≤–µ—Å—Ç–∏ –≤—Å–µ –∫ –≤–∏–¥—É –æ–¥–Ω–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞, –≤ –∫–æ—Ç–æ—Ä–æ–π –ø–æ–∫–∞–∑–∞–Ω–∞ —Ä–∞–∑–Ω–∏—Ü–∞, –∏ —Å–∫–∞–∑–∞—Ç—å, —á—Ç–æ —Ç–∞–º –≤—Å–µ —Ä–∞–≤–Ω–æ 0

c = []
for i in range(0,len(a)):
    c.append(a[i]-b[i])

result = ztest(c, value = 0, alternative = 'two-sided')

print(f'Z-statistics = {round(result[0],3)}\
        \n p-value = {round(result[1],3)}') # –≤—ã–≤–æ–¥–∏–º –∑–Ω–∞—á–µ–Ω–∏—è z-c—Ç–∞—Ç–∏—Å—Ç–∏–∫–∏ –∏ p-value, –æ–±–∞ –∑–Ω–∞—á–µ–Ω–∏—è –æ–∫—Ä—É–≥–ª—è–µ–º –¥–æ 3 –∑–Ω–∞–∫–æ–≤
```

# –ó–∞–¥–∞–Ω–∏–µ 17: –¥–≤—É—Ö–≤—ã–±–æ—Ä–æ—á–Ω—ã–π t-test

```python
group1 = [8.26, 18.15, 15.42, 17.95, 13.97,  9.89, 13.4 , 20.32, 24.91, 9.66]
group2 = [22.21, 16.22, 21.09, 15.23, 16.72, 18.34, 14.42, 15.39, 14.24, 18.41]
print(np.var(group1), np.var(group2)) #–≤—Å–ø–æ–º–∏–Ω–∞–µ–º –ø—Ä–∞–≤–∏–ª–æ 4:1
```

```python
result = stats.ttest_ind(a=group1, b=group2, equal_var=True)
print(f't-test = {round(result[0],3)}\
        \n p-value = {round(result[1],3)}')
```

```python
#31 –∑–∞–¥–∞–Ω–∏–µ
import statsmodels.api as sm
import numpy as np

# X - –º–∞—Ç—Ä–∏—Ü–∞ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ (–≤–∫–ª—é—á–∞—è JP_Sales^2 –∏ ln(1+NA_Sales))
# y - –º–∞—Å—Å–∏–≤ —Ü–µ–ª–µ–≤–æ–π –ø–µ—Ä–µ–º–µ–Ω–Ω–æ–π (Global_Sales)

# –°–æ–∑–¥–∞–Ω–∏–µ –º–∞—Ç—Ä–∏—Ü—ã –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ X
X = data['y']
# X = pd.concat([X, np.log1p(data['NA_Sales'])], axis=1)


# –î–æ–±–∞–≤–ª–µ–Ω–∏–µ –∫–æ–Ω—Å—Ç–∞–Ω—Ç–Ω–æ–≥–æ —Å—Ç–æ–ª–±—Ü–∞ –≤ –º–∞—Ç—Ä–∏—Ü—É –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ X
X = sm.add_constant(X)

# –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏ –ª–∏–Ω–µ–π–Ω–æ–π —Ä–µ–≥—Ä–µ—Å—Å–∏–∏
model = sm.OLS(data['price'], X)
results = model.fit()

# –ü–æ–ª—É—á–µ–Ω–∏–µ –æ—Ü–µ–Ω–æ–∫ –∫–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç–æ–≤
coeff = results.params
coeff
```

# –ó–∞–¥–∞–Ω–∏–µ 32

```python
mse = results.mse_resid
print("MSE –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ:", round(mse,2))
```

```python
from sklearn. metrics import mean_absolute_error as mae
data['Other_Sales_pred'] = coeff[0] + coeff[1] * data['JP_Sales'] + coeff[2] * data['NA_Sales']
actual = data['Other_Sales']
pred = data['Other_Sales_pred']
print("MAE –Ω–∞ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ:", round(mae(actual, pred),3))
```
